```python
import pandas as pd
```

## Load Data

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)[0]
```

## Filter not assigned Bourough

```python
df = df.loc[df.Borough != 'Not assigned', :]
```

## Group By and Concatenate Neighbourhood

```python
df.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(','.join).reset_index()
```

## Replace Values when neighborhood is not assigned

```python
df.loc[df.Neighbourhood == 'Not Assigned', ['Neighbourhood']] = df.loc[df.Neighbourhood == 'Not Assigned', ['Borough']]
```

```python
df.shape
```

## Load Geo Data

```python
geo_data = pd.read_csv('data/Geospatial_Coordinates.csv')
geo_data
```

```python
final_df = pd.merge(df, geo_data, how='inner', on='Postal Code')
```
